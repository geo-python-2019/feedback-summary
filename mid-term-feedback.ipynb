# Wordclouds!
In stead of highlighting individual quotes from the feedback, we tested how the feedback would look like as wordclouds :D 
A wordcloud summarizes a bunch of text, calculates word frequencies. The more popular the word, the bigger the text in the word cloud.

Word cloud plots in this notebook are based on this tutorial: https://www.datacamp.com/community/tutorials/wordcloud-python

## Preparations

This notebook and the related script file take advantage of the following modules
- numpy
- pandas
- matplotlib 
- pillow
- wordcloud 

The script file `wordcloudplotter.py` contains a function that generated a wordcloud based on a bunch of text.

Wordcloudplotter is fully based on this datacamp tutorial: https://www.datacamp.com/community/tutorials/wordcloud-python

In this notebook, we need to import only a couple of modules:

```python
import numpy as np
import pandas as pd
from wordcloudplotter import plot_word_cloud

import matplotlib.pyplot as plt
```
